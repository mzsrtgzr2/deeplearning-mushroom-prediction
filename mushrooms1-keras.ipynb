# mushrooms classifier deep learning

```python
import tensorflow as tf
import pandas as pd
import numpy as np

df = pd.read_csv('mushrooms.csv')
print(df.shape)
df.head()
#df.columns.values.tolist()
```

```python
# consider the folllowing features:
features = ['cap-shape',
             'cap-surface',
             'cap-color',
             'bruises',
             'odor',
             'gill-attachment',
             'gill-spacing',
             'gill-size',
             'gill-color',
             'stalk-shape',
             'stalk-root',
             'stalk-surface-above-ring',
             'stalk-surface-below-ring',
             'stalk-color-above-ring',
             'stalk-color-below-ring',
             'veil-type',
             'veil-color',
             'ring-number',
             'ring-type',
             'spore-print-color',
             'population',
             'habitat']

# encode the categorial series to numerators
from sklearn import preprocessing
lencoders = {}
for f in (features + ['class']):
    le = preprocessing.LabelEncoder()
    le.fit(df[f])
    df[f]=le.transform(df[f])
    lencoders[f] = le # keep label encoder
    
# select what you need from the DataFrame
data_features = df.ix[:,features] # take only features we want
data_classes = df['class']

data_classes_one_hot = []
for c in data_classes.values:
    if c == 0:
        data_classes_one_hot.append([1, 0])
    else:
        data_classes_one_hot.append([0, 1])
```

```python
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data_features, data_classes_one_hot, test_size=0.05, random_state=42)
X_train, X_valid, y_train, y_valid = train_test_split(X_train, y_train, test_size=0.05, random_state=43)
```

```python
from keras.models import Sequential

model = Sequential()

from keras.layers import Dense, Activation

model.add(Dense(units=128, input_dim=22))
model.add(Activation('relu'))
model.add(Dense(units=64))
model.add(Activation('relu'))
model.add(Dense(units=2))
model.add(Activation('softmax'))

model.compile(loss='categorical_crossentropy',
              optimizer='sgd',
              metrics=['accuracy'])
```

```python
# x_train and y_train are Numpy arrays --just like in the Scikit-Learn API.
model.fit(X_train.values, y_train, epochs=100, batch_size=64)
```
